Use concat for combining dataset on top of each other, your datasets are just stiched together along an axis. 

```python
pd.concat( [tb1,tb2] ) #concat directly merge one under the other(default) axis=0, join the dataset without taking common into account
```

```python
pd.concat( [tb1,tb2], ignore_index=True)
```

```python
pd.concat([tb1,tb2], axis=1) #ie stack side by side when axis=1 similar to merge but without common
```

```python
pd.concat([tb1,tb2], axis=1, join='inner') #join="outer"(default) ,coz nothing common so removed all NaN
```

```python
pd.concat([tb11,tb2], axis=1, join='inner') #tb11,tb2 axises are same
```

```python
pd.concat([tb11,tb2], axis=1, join='outer') 
```

```python
pd.concat([tb11,tb2], axis=0, join='inner')
```

```python
pd.concat([tb11,tb2], join='outer') #axis=0(default)
```
